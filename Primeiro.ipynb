```python
#Criando um projeto de cadastro de alunos e exibição de uma lista

cadastro = {}

class Aluno:
    def __init__(self, nome, nascimento, cpf):
        self.nome = nome

        # Processamento e atribuição de nascimento
        nascimento_str = str(nascimento)
        if len(nascimento_str) == 8:
            self.nascimento = f'{nascimento_str[:2]}/{nascimento_str[2:4]}/{nascimento_str[4:]}'
        else:
            print('Data de nascimento inválida')
            self.nascimento = nascimento_str # Atribui o valor original ou um padrão

        # Processamento e atribuição de cpf
        cpf_str = str(cpf)
        if len(cpf_str) == 11:
            self.cpf = f'{cpf_str[:3]}.{cpf_str[3:6]}.{cpf_str[6:9]}-{cpf_str[9:]}'
        else:
            print('CPF inválido')
            self.cpf = cpf_str # Atribui o valor original ou um padrão

        # Adiciona ao cadastro após processar nascimento e cpf
        cadastro[nome] = [self.nascimento, self.cpf]

    def __str__(self):
        return f"{self.nome} foi adicionado com sucesso.\nBem vindo! {self.nome}"

def listar():
    lista_alunos = list(cadastro.keys())
    lista_alunos.sort()
    print('Os alunos cadastrados são:')
    for idx, aluno_nome in enumerate(lista_alunos, start=1):
        # Exibindo os detalhes do aluno a partir do dicionário 'cadastro'
        detalhes = cadastro[aluno_nome]
        print(f'{idx}. Nome: {aluno_nome}, Nascimento: {detalhes[0]}, CPF: {detalhes[1]}')


opcao = input('Escolha uma das opções abaixo:\n1 - Adicionar aluno\n2 - Listar alunos\n3 - Sair\n')

while True:
    if opcao == '1':
        nome = input('Digite o nome do aluno:\n').title()
        nascimento = input('Digite a data de nascimento do aluno:\n')
        cpf = input('Digite o CPF do aluno:\n')
        aluno = Aluno(nome, nascimento, cpf)
        print(aluno) # Para ver a mensagem de sucesso
        opcao = input('Escolha uma das opções abaixo:\n1 - Adicionar aluno\n2 - Listar alunos\n3 - Sair\n')

    elif opcao == '2':
        listar()
        opcao = input('Escolha uma das opções abaixo:\n1 - Adicionar aluno\n2 - Listar alunos\n3 - Sair\n')

    elif opcao == '3':
        print('Saindo do programa...')
        break

print(cadastro)
```
